# Logistic Regression

```python
import numpy as np
import os
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay,plot_confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
def find_optimum_logisticCoeff(X_train,y_train,X_val,y_val, C_range):

    validation_scores = []
    train_scores = []

    for C in C_range:

        c_logit = LogisticRegression(C = C, random_state=0, max_iter=10000)

        c_logit.fit(X_train, y_train);

         # train on traning set
        train_scores.append(c_logit.score(X_train, y_train))
        # test on validation set
        validation_scores.append(c_logit.score(X_val,y_val))
    return train_scores,validation_scores
```

```python
X_train = np.loadtxt("Data_files/x_train.csv",delimiter=',')
y_train = np.loadtxt("Data_files/y_train.csv",delimiter=',', dtype=str)
X_val = np.loadtxt("Data_files/x_val.csv",delimiter=',')
y_val = np.loadtxt("Data_files/y_val.csv",delimiter=',', dtype=str)
X_test = np.loadtxt("Data_files/x_test.csv",delimiter=',')
y_test = np.loadtxt("Data_files/y_test.csv",delimiter=',', dtype=str)
```

```python
C_range = [1000, 100, 10, 1.0,0.1,0.01]
train_accuracy, validation_accuracy = find_optimum_logisticCoeff(X_train,y_train,X_val,y_val,C_range)
```

```python
plt.figure()
plt.plot(C_range, train_accuracy,label="Train Accuracy",marker='.')
plt.plot(C_range, validation_accuracy,label="Validation Accuracy",marker='.')
plt.xscale('log')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.legend()
plt.show();
```

```python
# Fitting the model with the optimum C value:

c_logit_optimum = LogisticRegression(C = 10, random_state=0, max_iter=10000)
c_logit_optimum.fit(X_train, y_train);

print(f'Score on train: {c_logit_optimum.score(X_train, y_train)}')
print(f'Score on validation: {c_logit_optimum.score(X_val, y_val)}')
print(f'Score on test: {c_logit_optimum.score(X_test, y_test)}')
```

# Logistic Regression with Scaled features

```python
X_train_scaled = np.loadtxt("Data_files/x_train_scaled.csv",delimiter=',')
y_train_scaled = np.loadtxt("Data_files/y_train_scaled.csv",delimiter=',', dtype=str)
X_val_scaled = np.loadtxt("Data_files/x_val_scaled.csv",delimiter=',')
y_val_scaled = np.loadtxt("Data_files/y_val_scaled.csv",delimiter=',', dtype=str)
X_test_scaled = np.loadtxt("Data_files/x_test_scaled.csv",delimiter=',')
y_test_scaled = np.loadtxt("Data_files/y_test_scaled.csv",delimiter=',', dtype=str)
```

```python
C_range = [1000, 100, 10, 1.0,0.1,0.01]
train_accuracy, validation_accuracy = find_optimum_logisticCoeff(X_train_scaled,y_train_scaled,X_val_scaled,y_val_scaled,C_range)
```

```python
plt.figure()
plt.plot(C_range, train_accuracy,label="Train Accuracy",marker='.')
plt.plot(C_range, validation_accuracy,label="Validation Accuracy",marker='.')
plt.xscale('log')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('EDA_plot/Logistic_train_validation.png')
plt.show();
```

```python
# Fitting the model with the optimum C value:

c_logit_optimum = LogisticRegression(C = 1, random_state=0, max_iter=10000)
c_logit_optimum.fit(X_train_scaled, y_train_scaled);

y_pred = c_logit_optimum.predict(X_test_scaled)
```

```python
cm_logistic = confusion_matrix(y_test_scaled, y_pred)
disp_logistic = ConfusionMatrixDisplay(confusion_matrix=cm_logistic, display_labels=['Angry', 'Calm', 'Fearful', 'Happy','Sad','Disgust','Surprised'])
fig, ax = plt.subplots(figsize=(8,8))
ax.set_title('Confusion Matrix for Logistic Regression')

disp_logistic.plot(xticks_rotation='vertical', ax=ax, cmap='BuGn')
plt.savefig('EDA_plot/logisticR_confusion_matrix.png')
plt.show()
```

```python
print(classification_report(y_test_scaled, y_pred))
```
